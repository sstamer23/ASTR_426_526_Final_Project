```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Retrieving
cut_1 = pd.read_csv("C:\\Users\\holnc\\Downloads\\Cut 2_4.csv", dtype = {"source_id": str})
cut_2 = pd.read_csv("C:\\Users\\holnc\\Downloads\\Cut 2_3.csv", dtype = {"source_id": str})
cutcut = pd.concat([cut_1, cut_2], ignore_index = True) #Putting both files together for simplicity, ignoring index for continuous counting.
ruwe = cutcut["ruwe"]

plt.figure(figsize = (10,10))
plt.hist(ruwe, bins = 100) #Using 100 bins here because it is easier to see that the peaks differentiate slightly at 1.

#Plotting
plt.axvline(1.4, c = 'r')
plt.xlabel("RUWE")
plt.ylabel("Number of Stars")
plt.title("RUWE Distribution")


plt.savefig("ASTR426_RUWEHist_FIX.png", dpi = 300)
plt.show()

#Filtering binaries for RUWE >= 1.4 for future use:
pot_binaries = cutcut[cutcut["ruwe"] > 1.4]
pot_binaries.to_csv("ASTR426_PotBin_FIX.csv")
```

```python
#Statistics:

#Exponential Fit:
ruwe = cutcut["ruwe"].values
ruwe = ruwe[~np.isnan(ruwe)]
counts, bins = np.histogram(ruwe, bins = 50)
bin_centers = (bins[:-1] + bins[1:]) / 2 

mask = counts > 0 #This is the moment where I realize the true value of masks.
x = bin_centers[mask]
y = np.log(counts[mask]) #Linearizing for numpy
m, b = np.polyfit(x, y, 1)

#y = Ae^Bx
A = np.exp(b)
B = m
x_fit = np.linspace(min(x), max(x), 200) #Out of 200 samples, take equal intervals of the min and max of x.
y_fit = A * np.exp(B * x_fit) #Exponentiate

plt.figure(figsize=(10, 6))
plt.hist(ruwe, bins = 50)
plt.plot(x_fit, y_fit)

plt.xlabel("RUWE")
plt.ylabel("Number of Stars")
plt.title("RUWE Histogram (Exponential Fit)")
plt.show()
print("Exponential Fit: ", A, B)

#Standard Deviation:
mean = np.mean(ruwe)
std = np.std(ruwe)
#Finding objects within 1 std:
sig1 = np.sum(( ruwe >= mean - std) & (ruwe <= mean + std))
tot = len(ruwe)

print("Standard Deviation: ", std)
print("Mean: ", mean)
print("W/in 1 std: ", sig1)
print("Percent of overall data: ", sig1/tot * 100, "%")
```
