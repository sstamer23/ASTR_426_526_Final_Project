```python
import matplotlib.pyplot as plt
import pandas as pd

file_gaia = 'Cut2_comb_12_non_ss_mag.csv'
df_gaia = pd.read_csv(file_gaia)
# ra_gaia = df_gaia['ra'].values
# dec_gaia = df_gaia['dec'].values
if df_gaia['phot_g_mean_mag'].values is not None:
    #print("G magnitude column exists in the Gaia dataset.")
    g_mag = df_gaia['phot_g_mean_mag'].values
    bp_mag = df_gaia['phot_bp_mean_mag'].values
    rp_mag = df_gaia['phot_rp_mean_mag'].values

colour = bp_mag - rp_mag

plt.figure(figsize=(10, 5))
plt.scatter(colour, g_mag, alpha=0.5, c='green', marker='o')
plt.xlabel('BP - RP (Colour Index)')
plt.ylabel('G Magnitude')
plt.title('HR diagram')
plt.gca().invert_yaxis()  # Invert y-axis for magnitude
plt.show()

```
